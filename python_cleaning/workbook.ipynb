```python
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('-', '_').str.lower()
```

```python
text_cols = df.select_dtypes(include=['object']).columns
for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
cols_to_fix = ['category', 'sub_category', 'city', 'state', 'region', 'segment', 'country', 'customer_name']
for col in cols_to_fix:
    df[col] = df[col].astype(str).str.title()
```

```python
num_cols = ['sales', 'quantity', 'discount', 'profit']
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df['postal_code'] = df['postal_code'].astype(str).str.zfill(5)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
df = df[df['sales'] > 0]
```

```python
df = df[df['profit'] <= df['sales']]
```

```python
df['days_to_ship'] = (df['ship_date'] - df['order_date']).dt.days
```

```python
errors = df[df['days_to_ship'] < 0]
print(f"Number of logic errors found: {len(errors)}")
```

```python
df['profit_margin'] = (df['profit'] / df['sales']) * 100
```

```python
extreme_cases = df[df['profit_margin'] > 100]
print(f"Found {len(extreme_cases)} rows with impossible margins.")
```

```python
df['order_year'] = df['order_date'].dt.year
df['order_month_name'] = df['order_date'].dt.month_name()
```

```python
# Check shape before dropping
before = df.shape[0]

# Drop duplicates
df = df.drop_duplicates(subset=['order_id', 'product_id'], keep='first')

# Check shape after dropping
after = df.shape[0]

print(f"Removed {before - after} duplicate rows.")
```

```python
import re

def has_hidden_chars(val):
    if isinstance(val, str):
        # This looks for anything outside the "Printable" ASCII range (32-126)
        if re.search(r'[^\x20-\x7E]', val):
            return True
    return False

# Check every column
for col in df.select_dtypes(include=['object']).columns:
    hidden_count = df[col].apply(has_hidden_chars).sum()
    if hidden_count > 0:
        print(f"Column '{col}' has {hidden_count} rows with hidden/non-standard characters.")
```

```python
import re

# Identify rows with non-standard characters in either column
mask = df['customer_name'].str.contains(r'[^\x20-\x7E]', na=False) | \
       df['product_name'].str.contains(r'[^\x20-\x7E]', na=False)

# Display the first 20 problematic rows
inspection_df = df[mask][['customer_name', 'product_name']].head(20)
print(inspection_df)
```

```python
import unicodedata

def final_data_polish(text):
    if not isinstance(text, str):
        return text
    
    # 1. Fix the non-breaking spaces (\xa0) from web-scraped data
    text = text.replace('\xa0', ' ')
    
    # 2. Normalize Accents (ö -> o, é -> e) 
    # This keeps the letter but removes the "hidden" complexity
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('ascii')
    
    # 3. Final safety check: Keep only standard printable characters
    # This automatically keeps +, /, -, ., (, ), and all numbers/letters
    # while removing smart quotes and other non-ASCII hidden symbols
    text = "".join(char for char in text if 31 < ord(char) < 127)
    
    return text.strip()

df['product_name'] = df['product_name'].apply(final_data_polish)
df['customer_name'] = df['customer_name'].apply(final_data_polish)
```

```python
import re

# Identify rows with non-standard characters in either column
mask = df['customer_name'].str.contains(r'[^\x20-\x7E]', na=False) | \
       df['product_name'].str.contains(r'[^\x20-\x7E]', na=False)

# Display the first 20 problematic rows
inspection_df = df[mask][['customer_name', 'product_name']].head(20)
print(inspection_df)
```

```python
final_dupes = df.duplicated(subset=['order_id', 'product_id']).sum()

if final_dupes > 0:
    print(f"Deep cleaning revealed {final_dupes} more duplicates. Removing them now...")
    df = df.drop_duplicates(subset=['order_id', 'product_id'], keep='first')
else:
    print("No additional duplicates found. The data is 100% unique per Product/Order.")
```
